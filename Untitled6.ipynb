```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.linear_model as lm
import numpy as np
# Úteis 
import datetime as dt
from datetime import timedelta

abast = pd.read_excel('abastpend.xlsx')

abast.rename(columns={'Combustível': 'Combus','Data(Mês)':'Data','Valor Total':'Valor' },inplace= True)

abast['Data'] = pd.to_datetime(abast['Data'])
abast
```

```python
####################################
# SEPARANDO OS COMBUSTIVEIS
#####################################
DS10 = abast.loc[abast.Combus == "DIESEL S-10"]
DS500 = abast.loc[abast.Combus == "DIESEL S-500"]
ET  = abast.loc[abast.Combus == "ETANOL COMUM"]
GAS = abast.loc[abast.Combus == "GASOLINA C ORIGINAL"]
GASADIT = abast.loc[abast.Combus == "GASOLINA ADITIVADA"]
#Converter DATE para datetime e associar ao index do dateframe

DS10.index = pd.to_datetime(DS10.Data, format="%m-%d-%Y")
DS500.index = pd.to_datetime(DS500.Data, format="%m-%d-%Y")
ET.index = pd.to_datetime(ET.Data, format="%m-%d-%Y")
GAS.index = pd.to_datetime(GAS.Data, format="%m-%d-%Y")
GASADIT.index = pd.to_datetime(GASADIT.Data, format="%m-%d-%Y")
#eliminar a coluna DATE, pois ela já seria o indice
DS10.drop("Data", inplace=True, axis=1)
DS500.drop("Data", inplace=True, axis=1)
ET.drop("Data", inplace=True, axis=1)
GAS.drop("Data", inplace=True, axis=1)
GASADIT.drop("Data", inplace=True, axis=1)

```
